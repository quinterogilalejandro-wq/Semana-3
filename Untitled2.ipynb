```python
import numpy as np

# 1. Funciones de Activación (Requisito 2)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriva(x):
    return x * (1 - x)

# 2. Dataset XOR - Clasificación Binaria (Requisito 3 y 4)
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# 3. Inicialización de Parámetros (Pesos y Sesgos)
np.random.seed(42) # Para resultados reproducibles
w_h = np.random.uniform(size=(2, 4)) # Capa oculta
w_o = np.random.uniform(size=(4, 1)) # Capa salida
b_h, b_o = np.zeros((1, 4)), np.zeros((1, 1))

# 4. Entrenamiento con Backpropagation (Requisito 1)
tasa_aprendizaje = 0.5
for epoca in range(10001):
    # Forward Propagation (Cálculo de salida y puntaje Z)
    capa_h = sigmoid(np.dot(X, w_h) + b_h)
    prediccion = sigmoid(np.dot(capa_h, w_o) + b_o)

    # Backpropagation (Cálculo del error y ajuste)
    error_o = y - prediccion
    d_salida = error_o * sigmoid_deriva(prediccion)

    error_h = d_salida.dot(w_o.T)
    d_oculto = error_h * sigmoid_deriva(capa_h)

    # Actualización de parámetros (Optimización con NumPy)
    w_o += capa_h.T.dot(d_salida) * tasa_aprendizaje
    w_h += X.T.dot(d_oculto) * tasa_aprendizaje

    if epoca % 2000 == 0:
        loss = np.mean(np.square(error_o))
        print(f"Época {epoca} | Error (Loss): {loss:.5f}")

# 5. Resultados Visibles (Requisito 4 y 5)
print("\n--- RESULTADOS FINALES DE CLASIFICACIÓN ---")
print(f"{'ENTRADA':<10} | {'META':<10} | {'PREDICCIÓN':<10}")
for i in range(len(X)):
    print(f"{str(X[i]):<10} | {str(y[i]):<10} | {prediccion[i][0]:.4f}")
```
