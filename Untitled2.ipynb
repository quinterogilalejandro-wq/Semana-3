```python
import numpy as np

# 1. FUNCIONES DE ACTIVACIÓN Y SUS DERIVADAS
# La derivada es esencial para Backpropagation (indica la dirección del ajuste)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriva(x):
    # Derivada de la sigmoide para el cálculo del gradiente
    return x * (1 - x)

# 2. MODELO DE RED NEURONAL MULTICAPA
class RedNeuronalEntrenable:
    def __init__(self, tam_entrada, tam_oculto, tam_salida):
        # Inicialización de parámetros (Pesos y Sesgos)
        self.w_oculto = np.random.uniform(size=(tam_entrada, tam_oculto))
        self.w_salida = np.random.uniform(size=(tam_oculto, tam_salida))
        self.b_oculto = np.zeros((1, tam_oculto))
        self.b_salida = np.zeros((1, tam_salida))

    def entrenar(self, X, y, epocas, tasa_aprendizaje):
        for epoca in range(epocas):
            # --- FORWARD PROPAGATION (Cálculo de salida) ---
            capa_oculta_z = np.dot(X, self.w_oculto) + self.b_oculto
            capa_oculta_a = sigmoid(capa_oculta_z)

            capa_salida_z = np.dot(capa_oculta_a, self.w_salida) + self.b_salida
            prediccion = sigmoid(capa_salida_z)

            # --- BACKPROPAGATION (Cálculo del error y ajuste) ---
            # 1. Error en la salida
            error_salida = y - prediccion
            d_salida = error_salida * sigmoid_deriva(prediccion)

            # 2. Error en la capa oculta (propagación hacia atrás)
            error_oculto = d_salida.dot(self.w_salida.T)
            d_oculto = error_oculto * sigmoid_deriva(capa_oculta_a)

            # --- ACTUALIZACIÓN DE PARÁMETROS ---
            self.w_salida += capa_oculta_a.T.dot(d_salida) * tasa_aprendizaje
            self.w_oculto += X.T.dot(d_oculto) * tasa_aprendizaje
            self.b_salida += np.sum(d_salida, axis=0, keepdims=True) * tasa_aprendizaje
            self.b_oculto += np.sum(d_oculto, axis=0, keepdims=True) * tasa_aprendizaje

            if epoca % 1000 == 0:
                loss = np.mean(np.square(error_salida)) # Error cuadrático medio
                print(f"Época {epoca} | Pérdida (Loss): {loss:.5f}")

# 3. DATASET DE CLASIFICACIÓN BINARIA (Problema XOR - No lineal)
# X = Entradas, y = Etiquetas reales
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]]) # La salida esperada

# 4. EJECUCIÓN DEL ENTRENAMIENTO
red = RedNeuronalEntrenable(tam_entrada=2, tam_oculto=4, tam_salida=1)
red.entrenar(X, y, epocas=10000, tasa_aprendizaje=0.1)

# 5. VERIFICACIÓN DE RESULTADOS
print("\n--- RESULTADOS FINALES ---")
for i in range(len(X)):
    # Predicción final tras el entrenamiento
    z_h = np.dot(X[i], red.w_oculto) + red.b_oculto
    a_h = sigmoid(z_h)
    z_o = np.dot(a_h, red.w_salida) + red.b_salida
    res = sigmoid(z_o)
    print(f"Entrada: {X[i]} | Meta: {y[i]} | Predicción: {res[0][0]:.4f}")
```
